# **Clase Matplottlib**

```python
df = pd.DataFrame(df)
print(df.info())
```

```python
# reemplazar la ,
#df['año'] = df['año'].str.replace(',','')
# transformar el valor a numerico
#df['año'] = pd.to_numeric(df['año'])

df['peso_nacer'] = pd.to_numeric(df['peso_nacer'])

print(df.info())
```

```python
# segmentar datos por ranking
#top = df.nlargest(2,['year_'])

#grafico de barras
plt.bar(df['year_'], df['edad_'])
plt.xlabel('year_')
plt.ylabel('edad_')
plt.title('edad por año')
```

```python
#grafico de barras
plt.bar(df['pac_hos_'], df['edad_'])
plt.xlabel('pac_hos_')
plt.ylabel('edad_')
plt.title('edad por año')
```

```python
#grafico de dispersion
plt.scatter(df['semana'], df['edad_'])
plt.xlabel('semana')
plt.ylabel('edad_')
plt.title('Dispercion')
plt.show()
```

```python
#Histograma
plt.hist(df['talla_nacer'], df['año_'])
plt.xlabel('talla_nacer')
plt.ylabel('año_')
plt.title('Histograma')
plt.show()
```

```python
#grafico de Torta

#plt.bar(df['year_'], df['edad_'])
#plt.pie(df['edad_'], df.labels=['id'], autopct='%1.1f%%')

plt.pie(df['year_'], autopct='%1.1f%%')
plt.title('Grafico de Torta')
plt.show()
```

```python
# Grafico de caja(boxplot)

#data = np.random.randn(100)

plt.boxplot(df['edad_'])
plt.ylabel('edad_')
plt.title('Grafico de cajas')
plt.show()
```

```python
# Grafico de violin

plt.violinplot(df['edad_'])
plt.title('Grafico de violin')
plt.show()
```

```python
# grafico de area

plt.fill_between(df['year_'],df['semana'])
plt.xlabel('year_')
plt.ylabel('semana')
plt.title('Grafico de area')
plt.show()
```

```python
# Grafico de barras apiladas

#plt.bar(categories, values_one, label = 'Grupo1')
#plt.bar(categories, values_two, bottom=values_one, label = 'Grupo2')
plt.bar(df['sexo_','pac_hos_'])
plt.xlabel('sexo_')
plt.ylabel('pac_hos_')
plt.title('Grafico de Barras apiladas')
plt.legend()
plt.show()
```

# Graficos **3D**

```python
import seaborn  as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
from matplotlib import cm
```

```python
# crear df
#np.random.seed(0)
#data = {'X': np.random.normal(0, 1, 1000), 'Y': np.random.normal(0, 1, 100)}
#df = pd.DataFrame(data)

# Grafico de densidades

sns.jointplot(x='X', y='Y', data=df, kind='kde')
plt.title('Grafico de densidades')
plt.show()
```

```python
# displot

#df = pd.DataFrame({'X': np.random.normal(0,1,1000)})

sns.displot(df['edad_'], kde= True)

plt.title('Displot')
plt.show()
```

```python
# grafico de pares
#df = pd.DataFrame(np.random.randn(100,4), columns =['A','B','C','D'])

sns.pairplot(df)
plt.title('Grafico de pares')

plt.show()
```

```python
# grafico de dispersion 3D

#df = pd.DataFrame(np.random.rand(100, 3), columns=['X','Y','Z'])
#fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
#ax.scatter(df['X'], df['Y'],df['Z'])
#ax.set_xlabel('X')
#ax.set_ylabel('Y')
#ax.set_zlabel('Z')
#plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['edad_'], df['edad_'],df['year_'])
ax.set_xlabel('edad_')
ax.set_ylabel('edad_')
ax.set_zlabel('year_')
plt.show()
```

```python
#grafico de lineas 3D
#ax.plot3D(df['X'], df['Y'],df['Z'], 'green')
#ax.set_xlabel('X')
#ax.set_ylabel('Y')
#ax.set_zlabel('Z')
#plt.show()

ax.plot3D(df['X'], df['Y'],df['Z'], 'green')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
